```python
# From NEPSE floorsheet, extract data of required pages and store it in a csv file.

from bs4 import BeautifulSoup
import requests
import pandas as pd

n = int(input("Enter page no = "))
final_data = []
for i in range(1,n+1):
    url = BeautifulSoup(f'http://www.nepalstock.com/main/floorsheet/index/{i}/', 'html.parser')
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data,'xml')
    
    tags = soup.table
    tag = tags.find_all('tr')
    if (i==1): #for first page, read table header
        tag = tag[1:-3]
    else: #for pages other than first, ignore table header
        tag = tag[2:-3]
    list_data = []   
    for i in range(len(tag)):
        list_data.append(tag[i].get_text())    
    for i in list_data:
        final_data.append(i.split('\n')[1:-1])

#write final_data to csv file
my_data_frame = pd.DataFrame(final_data)
my_data_frame.to_csv('floorsheet.csv',mode='w',header=False,index=None)

#Read from file and print it
df = pd.read_csv('floorsheet.csv',index_col='S.N.')
#df
```
