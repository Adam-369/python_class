```python
#Extract share price of required date and store it in relative csv file
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
from datetime import date,timedelta

date_extracted=date.today()-timedelta(1)
final_data = []   
i=1
my_url=f"http://www.nepalstock.com/main/todays_price/?startDate={date_extracted}&_limit=500"
    
#opening up connection,grabbing the page
uClient=uReq(my_url)
page_html=uClient.read()
uClient.close()

#html parsing
page_soup=soup(page_html,"html.parser")

#grab each traded companies data
data_table= page_soup.findAll("table",{"class":"table table-condensed table-hover"}) #Find the required table which has our data
data_rows=data_table[0].findAll("tr") #Find all rows inside that table

data_row=data_rows[1:-4] #Get the rows whose data is required 
for j in range(len(data_row)): #Processing the data inside each row as required to us
    temp_rowdata=(data_row[j].get_text()).split('\n') #Get text inside each row and split the data 
    temp_rowdata=[x for x in temp_rowdata if x] #Remove unnecessary spaces from the list of data
    final_data.append(temp_rowdata[0:10])    

#write final_data to csv file
my_data_frame = pd.DataFrame(final_data)
my_data_frame.to_csv(f'{date_extracted}_todaysprice.csv',mode='w',header=False,index=None)

#Read from file and print it
df = pd.read_csv(f'{date_extracted}_todaysprice.csv',index_col='S.N.')
df
```
